**Data Visulatization**

There are several functions in seaborn that allows you to visualize the data and explore various relationship between data features

```python
sns.countplot(data=df, x='Outcome')
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 4), tight_layout=True)
sns.boxplot(data=df)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 4), tight_layout=True)
sns.scatterplot(data=df, x='BMI', y ='DiabetesPedigreeFunction')
plt.show()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
sns.pairplot(df, vars=["Glucose","BloodPressure","BMI","Age"])
plt.show()
```

```python
sns.histplot(df["Age"])
plt.show()
```

```python
sns.histplot(df["BloodPressure"])
plt.show()
```

**Correlation between various features**

```python
df.corr(method='kendall')
```

**The heatmap helps to better see the relation between the features**

```python
sns.heatmap(df.corr(method='pearson'), annot=True)
plt.show()
```

```python
import numpy as np
#
# Correlation between different variables
#
corr = df.corr()
#
# Set up the matplotlib plot configuration
#
f, ax = plt.subplots(figsize=(10, 4))
#
# Generate a mask for upper traingle
#
mask = np.triu(np.ones_like(corr, dtype=bool))
#
# Configure a custom diverging colormap
#
cmap = sns.diverging_palette(230, 20, as_cmap=True)
#
# Draw the heatmap
#
sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```
